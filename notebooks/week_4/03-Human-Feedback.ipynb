### Import Dependencies

```python
from langsmith import Client
```

#### Attach Feedback to a specific trace in LangSmith

019c2ec2-59ad-72c0-8414-b5c146252675

```python
client = Client()
```

```python
RUN_ID = "019c2ec2-59ad-72c0-8414-b5c146252675"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by Aurimas",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by Aurimas",
    score=0,
    feedback_source_type="api"
)
```

```python
RUN_ID_RUN = "019c2ec2-59e9-7011-b2c0-28845ff9b43e"
```

```python
client.create_feedback(
    run_id=RUN_ID_RUN,
    key="comments_with_score",
    value="This is a long form free text feedback left by Aurimas",
    score=0,
    feedback_source_type="api"
)
```
